```python
import pickle
from transformers import *
```

```python
select_columns = ['n_runs',
 'total_km',
 'avg_km',
 'std_km',
 'med_pace_min_km',
 'std_pace_min_km',
 'avg_speed_kmh',
 'std_speed_kmh',
 'elev_gain_total',
 'elev_gain_avg',
 'elev_gain_std',
 'weekly_km']
```

```python
modelo = pickle.load(open('Downloads/output_modelo_2/model_rf.pkl', 'rb'))
```

```python
modelo
```

```python
tratamento_tabela = TratamentoTabela()
selector = Selector(selected_features=select_columns)
generate_score = GenerateScore(modelo=modelo)
```

```python
INPUT_CSV  = r"C:\Users\gtvca\Downloads\raw-data-kaggle (3).csv"
DELIMITER  = ";"  # Se o seu CSV for separado por ;
df = pd.read_csv(INPUT_CSV, delimiter=DELIMITER)
```

```python
df_teste = df[df['athlete']==7947798]
```

```python
X = tratamento_tabela.transform(df_teste)

X = selector.transform(X)

X = generate_score.predict(X)
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([('tratamento_tabela', tratamento_tabela), ('selector', selector), ('generate_score', generate_score)])
```

```python
pipe
```

```python
pickle.dump(pipe, open('Downloads/output_modelo_2/pipeline_modelo.pkl', 'wb'))
```
